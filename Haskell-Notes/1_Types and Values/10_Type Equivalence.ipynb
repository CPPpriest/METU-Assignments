## "Name Equivalence"  
  
Haskell uses Name Equivalence:  
Data types that are defined on the same place, with same name are EQUIVALENT.  

```haskell
data Tree a = ETree | Node( a, Tree a, Tree a)

type BRT = Tree Bool 

x = Node( True, ETree, ETree )
y = Node( True, ETree, ETree ) :: BRT

:t x
:t y
```

Defining y as BRT type is Legal due to type Equivalence  
'type' keyword just creates an alias to the same data type

## "Structural Equivalence"   
  
Haskell do NOT have Structural Type Equivalence !!!

```haskell
data Tree a = ETree | Node( a, Tree a, Tree a)
data HTree a = HETree | HNode( a, HTree a, HTree a)

x = Node( True, ETree, ETree )
:t x

y = Node( True, ETree, ETree ) :: HTree
:t y
```

Although Tree and HTree have structural equivalence (Same set of values),  
Throws an exception: UNMATCHING TYPES
