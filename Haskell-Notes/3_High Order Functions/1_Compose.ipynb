# Higher Order Function
Functions that do at least one of the following:  
  
1) **Take one or more functions as arguments:**    
    A higher-order function can accept other functions as parameters, allowing for flexible and reusable code patterns.  
    This can be used to abstract or generalize behavior that can be customized with different functions.
    
3) **Return another function as its result:**    
    Higher-order functions can also return functions as their outputs.  
    This allows for creating functions dynamically, based on conditions or inputs during runtime.

```haskell
f x = x + x
g x = x * x

compose1 func1 func2 x = func1 (func2 x)
compose2 func1 func2 x = func1 $ func2 x
compose3 func1 func2 = func1 . func2

compose1 f g 4
compose2 f g 4
compose3 f g 4
```

All the compose functions do the same thing: make g x executed first and f is called with the returned value.  
In the context, f is taking another function as its argument: Higher Order Function

```haskell
:t compose1
```

**compose: (a -> b) -> (c-> a) -> c -> b**  
**func1: (a -> b) -> a** ___Takes a and returns b    
**func2: (c -> a) -> c** ___Takes c and returns a  


```haskell
u f g = f . g
:t u

w = f . g
:t w
```

u is a function that takes two function arguments:  
**(b -> c) -> (a -> b) -> a -> c**  
  
w is a new function that is defined as composition:  
**a -> c**

Realize that the first two functions (As arguments) in u definition is not included in w definition.  
Last part is exactly the same.
