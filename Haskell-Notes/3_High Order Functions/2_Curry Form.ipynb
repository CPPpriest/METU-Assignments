# Cartesian Form

The Cartesian form (or uncurried form) of a function takes multiple arguments simultaneously, usually as a tuple or a set of parameters.   
Function is applied to all its arguments at once, rather than returning a new function after each argument.

```haskell
add :: (Int, Int) -> Int
add (x, y) = x + y

:t add
```

# Curried Form

Currying is a technique where a function with multiple arguments is transformed into a sequence of functions, each with a single argument.  
In essence, a curried function takes one argument and returns another function expecting the next argument, and so on.   
This continues until all arguments have been provided, at which point the final value is computed.

```haskell
add :: Int -> Int -> Int
add x y = x + y

:t add
```

# "Curry" Function in Haskell

A standard library function that converts a functiocurry f x y = f (x, y) from its Cartesian form to its curried form.

```haskell
:t curry
```

```haskell
func (x,y) = x + y    -- Cartesian Form
func(3,4)
:t func

func2 = curry func    -- Curried Form
func2 3 4
:t func2
```

**Example:**

```haskell
add (x,y) = x + y
:t add

increment = curry add 1
:t increment

increment 5
```
