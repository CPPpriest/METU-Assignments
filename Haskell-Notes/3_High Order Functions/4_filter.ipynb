# filter


1) Gets a **Boolean** function and a list
2) Applies the function to all elements and keeps the **True** elements
3) Drops the **False** returning elements (filters)
4) Returns the new (filtered) list

```haskell
map :: (a -> Bool) -> [a] -> [a]

filter boolFunc [] = []
filter boolFunc (el : rest) =  if (boolFunc el) 
                                  then el : (filter boolFunc rest) 
                               else (filter boolFunc rest)


```

````haskell
isEven x = x `mod` 2 == 0

filter isEven [0..10]
````
